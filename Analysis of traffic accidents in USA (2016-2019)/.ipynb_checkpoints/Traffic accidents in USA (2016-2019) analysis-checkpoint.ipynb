#### h0: there is no correlation between the number of traffic accidents and the time of the day. 
#### h1: there is a positive correlation between the number of accidents and the time of the day.

```python
accident_df = df2[['ID', 'Source', 'TMC', 'Severity', 'Start_Time', 'End_Time',
       'Start_Lat', 'Start_Lng', 'Distance(mi)',
       'Description', 'Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit',
       'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming',
       'Traffic_Signal', 'Turning_Loop', 'Sunrise_Sunset',
       'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight']]
```

```python
accident_df['Start_Time'].value_counts()
```

```python
# Taking 30% sample from the data
sample_start_time = accident_df['Start_Time'].sample(frac = 0.3, random_state=1) 
```

```python
counter_st = Counter(sample_start_time)
```

```python
counter_st_plot = pd.DataFrame.from_dict(counter_st, orient='index')
```

```python
counter_st_plot.plot(figsize=(30, 10));
```

On the graph we can clearly see that distribution of data has two picks: time from 6 to 9 am and time from 4 to 6 pm. Time when people usually go to work and return from it. Which means that H1 hypothesis is correct.

#### As conclusion I would not recommend to drive a car from 6 to 9 am and from 4 to 6 pm in Baltimore county, Richmond City and South Carolina as a whole, especially with high humidity.
