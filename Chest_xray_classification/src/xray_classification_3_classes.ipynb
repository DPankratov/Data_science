```python
import os
import cv2
import numpy as np 
import pandas as pd 

import seaborn as sns 
sns.set_style('darkgrid')
import matplotlib.pyplot as plt
plt.style.use('dark_background')
from PIL import Image, ImageDraw 
from pylab import savefig

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, AvgPool2D, MaxPool2D , Flatten , Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
IMAGE_SIZE = [150, 150]
EPOCHS = 100
BATCH_SIZE = 32
```

Pictures have different sizes and needed to be scaled

Let's see how our data is distributed

```python
labels = ['PNEUMONIA', 'NORMAL']

def get_training_data(data_dir):
    data = [] 
    for label in labels: 
        path = os.path.join(data_dir, label)
        class_num = labels.index(label)
        matrix = ""
        
        for img in os.listdir(path):
            img_arr = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            resized_arr = cv2.resize(img_arr, (IMAGE_SIZE[0], IMAGE_SIZE[1])) # Reshaping images to preferred size
            if "bacteria" in img:
                matrix = [1,0,0] 
            elif "virus" in img:
                matrix = [0,1,0] 
            else:
                matrix = [0,0,1]              
            data.append([resized_arr, class_num, matrix])

    return np.array(data)
```

```python
train = get_training_data('../input/chest_xray/train')
val = get_training_data('../input/chest_xray/val')
test = get_training_data('../input/chest_xray/test')
```

```python
#visualization of data by category of images
def display_set(one_set, order = ["bacteria", "virus", "normal"]):
    l = []
    for i in one_set:
        if i[2] == [1,0,0]:
            matrix = "bacteria"
        elif i[2] == [0,1,0]:
            matrix = "virus"
        else:
            matrix = "normal"
        l.append(matrix)
    svm = sns.countplot(l, order = order).set_title('Train data distribution');
    figure = svm.get_figure()    
    figure.savefig('train_distr.png', dpi=200)
```

Data in categories is imbalanced

```python
# plot all data based on the category
x_data = []
y_data = []

for feature, label, matrix in train:
    x_data.append(feature)
    y_data.append(matrix)

for feature, label, diag in test:
    x_data.append(feature)
    y_data.append(matrix)
    
for feature, label, diag in val:
    x_data.append(feature)
    y_data.append(matrix)

y_class_num = [np.where(np.asarray(r)==1)[0][0] for r in y_data]
svm = sns.countplot(y_class_num).set_title('All data');
figure = svm.get_figure()    
figure.savefig('svm_conf.png', dpi=200)
```

Data is imbalances and it needs to be corrected. Now we will even the number of examples of categories.

## Data preparation

```python
# Undersampling the imbalanced data

samples = pd.DataFrame(y_class_num, columns = ['category'])
samples['matrix'] = y_data
samples['img'] = x_data

sm_matrix_count = samples['category'].value_counts().min() #what is the minimum of samples among the categories
print("Number of samples in category", sm_matrix_count);

class0 = samples[samples['category'] == 0].sample(sm_matrix_count) # undersampling all categories based on the minimum
class1 = samples[samples['category'] == 1].sample(sm_matrix_count)
class2 = samples[samples['category'] == 2].sample(sm_matrix_count)

samples_under = pd.concat([class0, class1, class2], axis=0)

x = samples_under['img'].tolist()
y = samples_under['matrix'].tolist()
svm = sns.countplot(samples_under['category']).set_title('Final Test data');
figure = svm.get_figure()    
figure.savefig('final_test_data.png', dpi=200)
```

Lets make a new split and shuffle the data (80 / 10 / 10)

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True) # train and the rest
svm = sns.countplot([np.where(np.asarray(r)==1)[0][0] for r in y_train]).set_title('Training data');
figure = svm.get_figure()    
figure.savefig('train_data_after_undersampling.png', dpi=200)
```

```python
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=0.5, shuffle=False) # validation and test data 
svm = sns.countplot([np.where(np.asarray(r)==1)[0][0] for r in y_val]).set_title('Validation data');
figure = svm.get_figure()    
figure.savefig('val_data_after_undersampling.png', dpi=200)
```

```python
svm = sns.countplot([np.where(np.asarray(r)==1)[0][0] for r in y_test]).set_title('Test data');
figure = svm.get_figure()    
figure.savefig('test_data_after_undersampling.png', dpi=200)
```

Data normalization

```python
x_train = np.array(x_train) / 255
x_val = np.array(x_val) / 255
x_test = np.array(x_test) / 255
```

Data scaling

```python
x_train = x_train.reshape(-1, IMAGE_SIZE[0], IMAGE_SIZE[1], 1)
y_train = np.array(y_train)

x_val = x_val.reshape(-1, IMAGE_SIZE[0], IMAGE_SIZE[1], 1)
y_val = np.array(y_val)

x_test = x_test.reshape(-1, IMAGE_SIZE[0], IMAGE_SIZE[1], 1)
y_test = np.array(y_test)
```

Data augmentation in order to avoid overfitting

```python
train_datagen = ImageDataGenerator(
        featurewise_center = False,  # set input mean to 0 over the dataset
        samplewise_center = False,  # set each sample mean to 0
        featurewise_std_normalization = False,  # divide inputs by std of the dataset
        rotation_range = 10,  # randomly rotate images in the range (degrees, 0 to 180)
        zoom_range = 0.1, # Randomly zoom image 
        width_shift_range = 0.1,  # randomly shift images horizontally (fraction of total width)
        height_shift_range = 0.1,  # randomly shift images vertically (fraction of total height)
        horizontal_flip = False,  # randomly flip images
        vertical_flip = False  # randomly flip images
        ) 

train_datagen.fit(x_train)

valid_datagen = ImageDataGenerator()
test_datagen = ImageDataGenerator()

train_generator = train_datagen.flow(
    x_train,
    y_train,
    batch_size = BATCH_SIZE)

validation_generator = valid_datagen.flow(
    x_val,
    y_val)
```

## Building CNN model

```python
def build_model():
    model = Sequential()
    model.add(Conv2D(32 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu' , input_shape = (IMAGE_SIZE[0], IMAGE_SIZE[1], 1)))
    model.add(AvgPool2D((2,2) , strides = 2 , padding = 'same'))

    model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
    model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))

    model.add(Conv2D(64 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
    model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))

    model.add(Conv2D(128 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
    model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))

    model.add(Conv2D(256 , (3,3) , strides = 1 , padding = 'same' , activation = 'relu'))
    model.add(MaxPool2D((2,2) , strides = 2 , padding = 'same'))

    model.add(Flatten())
    model.add(Dense(units = 128 , activation = 'relu'))
    model.add(Dense(units = 3 , activation = 'softmax'))
    
    return model
```

## Training the model

```python
early_stop = tf.keras.callbacks.EarlyStopping(
    patience = 10, restore_best_weights = True
)

initial_learning_rate = 1e-3
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate, decay_steps = 100000, decay_rate = 0.96, staircase = True
)
```

```python
model = build_model()

METRICS = [
    'accuracy',
    tf.keras.metrics.Precision(name = "precision"),
    tf.keras.metrics.Recall(name = "recall"),
    tf.keras.metrics.AUC(name = "AUC"),
    ]

model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate = lr_schedule),
    loss = "categorical_crossentropy",
    metrics = METRICS,
    )

history = model.fit(
    train_generator,
    epochs = EPOCHS,
    validation_data = validation_generator,
    callbacks = [early_stop],
)
```

## Performance

```python
fig, ax = plt.subplots(1, 5, figsize=(30, 4))
ax = ax.ravel()
plt.style.use('dark_background')

for i, j in enumerate(["precision", "recall", "accuracy", "AUC", "loss"]):
    ax[i].plot(history.history[j])
    ax[i].plot(history.history["val_" + j])
    ax[i].set_title("Model {}".format(j))
    ax[i].set_xlabel("epochs")
    ax[i].set_ylabel(j)
    ax[i].legend(["train", "val"])
fig.savefig('performance.png', dpi=200)
```

```python
print("Loss of the model is - " , model.evaluate(x_test, y_test)[0]*100)
print("Accuracy of the model is - " , model.evaluate(x_test, y_test)[1]*100 , "%")
```

```python
predictions = model.predict_classes(x_test)
y_test_num = [np.where(r==1)[0][0] for r in y_test]
names = ['Pneumonia Bacteria', 'Pneumonia Virus','Normal']

print(classification_report(y_test_num, predictions, target_names = names));
```

```python
cm = confusion_matrix(y_test_num, predictions)
cm = pd.DataFrame(cm , index = names , columns = names)
cm.index.name = 'Actual'
cm.columns.name = 'Predicted'

group_counts = ["{0:0.0f}".format(value) for value in cm.to_numpy().flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cm.to_numpy().flatten()/np.sum(cm.to_numpy())]
labels = [f"{v1}\n{v2}" for v1, v2 in zip(group_counts,group_percentages)]
labels = np.asarray(labels).reshape(3,3)

plt.figure(figsize = (12,12))
svm = sns.heatmap(cm,
            annot=labels,
            cmap= "coolwarm",
            linecolor = 'black',
            linewidth = 1,
            fmt=''
           );
figure = svm.get_figure()    
figure.savefig('confusion_matrix.png', dpi=400)
```

Our model shows good results distinguishing Normal VS Pneumonia images. Number of False Negative is minimum which is very good. However, results for distinguishing Bacterial VS Virus images could have been better.
