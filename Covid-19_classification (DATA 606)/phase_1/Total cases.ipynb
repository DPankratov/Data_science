```python
import seaborn as sns
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 10,5

sns.set(style="ticks", context="talk")
plt.style.use("dark_background")
series = pd.read_csv('Total confirmed cases.csv', header=0, index_col=0, parse_dates=True, squeeze=False)
```

```python
ax = series.plot()
ax.set_title('Total COVID-19 cases');
ax.set_xlabel('');
ax.set_ylabel('Cases');
plt.savefig('cases.png')
```

```python
series2 = pd.read_csv('Global totals.csv', header=0, index_col=0, parse_dates=True, squeeze=False)
```

```python
ax = series2.plot()
ax.set_title('Recovered / infected / deaths');
ax.set_xlabel('');
ax.set_ylabel('Cases');
plt.savefig('recovered.png')
```
